# Synthetic datasets
This notebook creates synthetic datasets mimiking gene-expression data. It is mainly devised to test the SVDmerge method.

```python
import numpy as np
import pandas as pd
```

```python
def rand_gauss_dataset(means_mean,variances_mean,N_genes,N_patients):
    means = np.random.normal(loc=means_mean,size=N_genes)
    variances = [x**2 for x in np.random.normal(loc=variances_mean,size=N_genes)]
    data = np.array([np.random.normal(loc=m,scale=s,size=N_patients) for m,s in zip(means,variances) ])
    df = pd.DataFrame(
        data =  data.T,
        index = ["patient"+str(i) for i in range(N_patients)],
        columns = ["gene"+str(i) for i in range(N_genes)]
    )
    return df
```

```python
batch1 = rand_gauss_dataset(means_mean=6,variances_mean=1,N_genes=20,N_patients=14)
```
